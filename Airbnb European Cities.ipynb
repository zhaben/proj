# Airbnb European Cities

Dataset: https://www.kaggle.com/datasets/thedevastator/airbnb-prices-in-european-cities?resource=download

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import os
os_files = os.listdir("/Users/zhanebennett/Downloads/archive")
```

```python
dfs = []
path = "/Users/zhanebennett/Downloads/archive"

for file in os_files:
    df = pd.read_csv(os.path.join(path, file))
    df['City'] = file.split('_')[0]
    df['Time_of_week'] = file.split('_')[1][:-4]
    dfs.append(df) #list of dataframes
    
```

```python
df_combined = pd.concat(dfs)
```

```python
df_combined.drop(columns=df_combined.columns[0], axis=1, inplace=True)
```

```python
df = df_combined.copy()
```

## Manage Outliers

```python
#Outliers

outliers = df[['realSum']].dropna()
df_outliers = df.loc[outliers.apply(lambda x: np.abs(x - x.mean()) / x.std() < 3).all(axis=1)]
#outliers2.apply(lambda x: np.abs(x - x.mean()) / x.std() < 3).all(axis=1)
df_outliers.head()
```

```python
df_outliers.count()
```

```python
# Dataframe is now without outliers
df = df_outliers.copy()
```

```python
normalized_df=(df-df.mean())/df.std() #mean
#normalized_df=(df-df.min())/(df.max()-df.min()) #max_min
```
